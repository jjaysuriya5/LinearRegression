```python
x = ip_data.iloc[ : , :-1].values
y = ip_data.iloc[ : , -1].values
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y ,
                                                      test_size = 0.25 , random_state = 0)
```

```python
reg = LinearRegression()
reg.fit(x_train , y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
s = ''
for i in range(0,len(reg.coef_)):
    s = s + ' {} X{} + '.format(str(reg.coef_[i]) , i)
s = s.rstrip('+ ')
print('The linear equation is : Y = {1} + {0}'.format(s,reg.intercept_))
```

```python
# showing the training data set
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_train[ : , 0] , x_train[ : ,1] , y_train , color ='red' )
# ax.plot(x_train[ : , 0] , x_train[ : ,1] , reg.predict(x_train)  )
plt.show()
```

```python
y_pred =  reg.predict(x_test)
mse = mean_squared_error( y_test , y_pred )
rmse = np.sqrt(mse)

r2 = r2_score( y_test , y_pred )

print('The mean square error is : ' , mse )
print('The root mean square error is : ' , rmse )
print('The R2 value is : ' , r2 )
```

```python
# to check if my regression model is correct by plotting residual graph
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter( x_test[ : , 0] , x_test[ : ,1] , y_test - y_pred , color = 'red')
ax.plot( x_test[ : , 0] , x_test[ : ,1] ,  x_test[ : ,1]*0 )
```

```python
# to check if my regression model is correct by plotting residual graph
fig = plt.figure()
x1 = [i for i in range( 1 , len(y_pred)+1)]
plt.scatter( x1 , y_test - y_pred )
plt.plot(x1,y_test*0 ,color = 'red')
```
